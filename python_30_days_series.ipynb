## Day 5 : Lists - Indexing, slicing, methods, list comprehension


A list in Python is an ordered, mutable collection of elements.

You can store different data types in a list (integers, strings, floats, even other lists).

So, when you add a new item it will be saved at the end of the list.

You can change, add, or remove elements after creation.



#When NOT to Use Lists

âŒ If you only need unique values â†’ use set.

âŒ If you need key-value pairs â†’ use dict.

âŒ If data should be unchangeable â†’ use tuple.

# To add value and element in list

1. Append - Adds an element at the end of the list
2. insert - Adds an element at the specified position
3. extend - Add the elements of a list (or any iterable), to the end of the current list

```python
# Insert
fruits = ['apple', 'banana', 'cherry']

fruits.insert(1, "orange")
print(fruits)
```

```python
# extend
fruits = ['apple', 'banana', 'cherry']

points = (1, 4, 5, 9)

fruits.extend(points)
print(fruits)
```

# To delete an element from list

1. remove - Removes the item with the specified value
2. pop - Removes the element at the specified position
3. clear - Removes all the elements from the list

```python
# remove
fruits = ['apple', 'banana', 'cherry']

fruits.remove("banana")
print(fruits)
```

```python
# pop
fruits = ['apple', 'banana', 'cherry']

fruits.pop(1)
print(fruits)
```

# If - else structure

In a list comprehension, there are two different kinds of if:

Filter if (at the end):
[expr for item in iterable if condition]

Conditional Expression (ifâ€“else) (inside expression)
[expr_if_true if condition else expr_if_false for item in iterable]

```python
#Given this list of dictionaries representing products:
#pythonproducts = [
#    {"name": "laptop", "price": 1000, "category": "electronics"},
 #   {"name": "chair", "price": 150, "category": "furniture"},
  #  {"name": "phone", "price": 800, "category": "electronics"},
   # {"name": "desk", "price": 300, "category": "furniture"}]
#Group them by category and calculate the average price for each category.

products = [
    {"name": "laptop", "price": 1000, "category": "electronics"},
    {"name": "chair", "price": 150, "category": "furniture"},
    {"name": "phone", "price": 800, "category": "electronics"},
    {"name": "desk", "price": 300, "category": "furniture"}
]

# Group prices by category
category_prices = {}

for product in products:
    category = product["category"]
    price = product["price"]

    if category not in category_prices:
        category_prices[category] = []

    category_prices[category].append(price)

# Calculate averages
for category, prices in category_prices.items():
    average = sum(prices) / len(prices)
    print(f"{category}: Average price = ${average:.2f}")

```

```python
# Write a program that simulates a complex traffic light system with pedestrian
# crossing, emergency vehicle priority, and rush hour timing. %%

import time
from datetime import datetime

def get_current_hour():
    """Get current hour or simulate for testing"""
    # For testing, you can change this value
    current_time = datetime.now().hour
    return current_time

def is_rush_hour(hour):
    """Determine if current time is rush hour"""
    morning_rush = 7 <= hour <= 9
    evening_rush = 17 <= hour <= 19  # 5 PM to 7 PM
    return morning_rush or evening_rush

def is_night_time(hour):
    """Determine if it's night time"""
    return hour >= 22 or hour <= 5  # 10 PM to 5 AM

def get_traffic_timing(hour):
    """Get traffic light timing based on time of day"""
    if is_night_time(hour):
        return {
            'green': 15,
            'yellow': 3,
            'red': 15,
            'pedestrian_walk': 10
        }
    elif is_rush_hour(hour):
        return {
            'green': 60,     # Extended green during rush hour
            'yellow': 5,
            'red': 25,
            'pedestrian_walk': 20
        }
    else:
        return {
            'green': 30,
            'yellow': 5,
            'red': 25,
            'pedestrian_walk': 15
        }

def handle_emergency_override():
    """Handle emergency vehicle priority"""
    print("ðŸš¨ EMERGENCY VEHICLE DETECTED! ðŸš¨")
    print("Immediately switching to GREEN for emergency vehicle")
    print("Emergency green light: 45 seconds")
    print("Emergency vehicle passed. Resuming normal cycle...")
    return True

def handle_pedestrian_crossing(timing):
    """Handle pedestrian crossing request"""
    print(f"\nðŸ‘¥ Pedestrian crossing activated")
    print(f"ðŸš¶ WALK signal: {timing['pedestrian_walk']} seconds")
    print("ðŸš¶ DON'T WALK (flashing): 5 seconds")
    print("Pedestrians can cross safely")

def display_traffic_state(state, duration, additional_info=""):
    """Display current traffic light state"""
    colors = {
        'green': 'ðŸŸ¢ GREEN',
        'yellow': 'ðŸŸ¡ YELLOW',
        'red': 'ðŸ”´ RED'
    }
    print(f"{colors[state]} light: {duration} seconds {additional_info}")

def simulate_traffic_light():
    """Main traffic light simulation"""
    print("ðŸš¦ COMPLEX TRAFFIC LIGHT SYSTEM ðŸš¦")
    print("=" * 50)

    # Get inputs
    current_hour = int(input("Enter current hour (0-23): "))

    print(f"\nCurrent time: {current_hour}:00")

    # Determine time-based conditions
    if is_night_time(current_hour):
        print("ðŸŒ™ NIGHT MODE ACTIVATED")
        print("ðŸ”´ Blinking red lights - Stop and proceed when safe")
        return

    timing = get_traffic_timing(current_hour)

    if is_rush_hour(current_hour):
        print("ðŸš— RUSH HOUR MODE - Extended green timing")
    else:
        print("ðŸ• NORMAL HOURS - Standard timing")

    # Check for emergency vehicle
    emergency = input("\nIs emergency vehicle approaching? (yes/no): ").lower() == 'yes'

    if emergency:
        handle_emergency_override()
        print("\nResuming normal cycle after emergency...")

    # Check for pedestrian request
    pedestrian_request = input("Pedestrian crossing button pressed? (yes/no): ").lower() == 'yes'

    print("\n" + "="*50)
    print("TRAFFIC LIGHT CYCLE:")
    print("="*50)

    # Normal traffic cycle
    print(f"\nðŸš¦ Current timing pattern:")
    display_traffic_state('green', timing['green'])

    if pedestrian_request:
        print("â³ Pedestrian request noted - will activate after vehicle green phase")

    display_traffic_state('yellow', timing['yellow'], "(Prepare to stop)")
    display_traffic_state('red', timing['red'])

    # Handle pedestrian crossing during red phase
    if pedestrian_request:
        handle_pedestrian_crossing(timing)

    print(f"\nðŸ“Š CYCLE SUMMARY:")
    print(f"Total cycle time: {timing['green'] + timing['yellow'] + timing['red']} seconds")
    print(f"Time of day: {'Rush Hour' if is_rush_hour(current_hour) else 'Normal Hours'}")
    print(f"Emergency override: {'Yes' if emergency else 'No'}")
    print(f"Pedestrian crossing: {'Activated' if pedestrian_request else 'Not requested'}")

def advanced_scenario_demo():
    """Demonstrate complex scenarios"""
    print("\n" + "="*60)
    print("ðŸŽ¯ ADVANCED SCENARIOS DEMONSTRATION")
    print("="*60)

    scenarios = [
        {
            'hour': 8, 'emergency': True, 'pedestrian': True,
            'description': 'Rush Hour + Emergency + Pedestrian'
        },
        {
            'hour': 14, 'emergency': False, 'pedestrian': True,
            'description': 'Normal Hours + Pedestrian Only'
        },
        {
            'hour': 23, 'emergency': False, 'pedestrian': False,
            'description': 'Night Time Mode'
        }
    ]

    for i, scenario in enumerate(scenarios, 1):
        print(f"\n--- SCENARIO {i}: {scenario['description']} ---")
        hour = scenario['hour']

        if is_night_time(hour):
            print("ðŸŒ™ Night mode: Blinking red lights")
            continue

        timing = get_traffic_timing(hour)

        if scenario['emergency']:
            print("ðŸš¨ Emergency vehicle priority activated!")
            print("âš¡ Immediate green: 45 seconds")

        print(f"ðŸŸ¢ Green: {timing['green']}s")
        print(f"ðŸŸ¡ Yellow: {timing['yellow']}s")
        print(f"ðŸ”´ Red: {timing['red']}s")

        if scenario['pedestrian']:
            print(f"ðŸš¶ Pedestrian walk: {timing['pedestrian_walk']}s")

# Run the simulation
if __name__ == "__main__":
    try:
        simulate_traffic_light()

        demo = input("\nWould you like to see advanced scenarios demo? (yes/no): ")
        if demo.lower() == 'yes':
            advanced_scenario_demo()

    except ValueError:
        print("âŒ Please enter valid numeric values!")
    except KeyboardInterrupt:
        print("\n\nðŸ›‘ Traffic light system stopped!")
```

## Encapsulation In Python
Encapsulation is a fundamental principle in object-oriented programming that focuses on bundling data and the methods that operate on that data into a single unit called a class. It allows you to control the access and visibility of the data and methods, providing a way to protect and organize your code

```python
### Access Modifier--->Encapsulation
## private

class Person:
  def __init__(self,name,age):
    self.__name = name
    self.__age = age

  def display_info(self):
        print(f"the person name is {self.__name} and the age is {self.__age}")
```

```python
## Rock paper & Scissor game
import random

given_choice = ['Rock','Paper','Scissor']
user_choice = input("Enter your choice (Rock, Paper, Scissor): ")
computer_choice = random.choice(given_choice)

while user_choice != 'quit':
  if user_choice == 'quit':
    break
  elif user_choice == computer_choice:
    print("It's a tie!")
  elif user_choice == 'Rock' and computer_choice == 'Scissor':
    print("You win!")
  elif user_choice == 'Scissor' and computer_choice == 'Paper':
    print("You win!")
  elif user_choice == 'Paper' and computer_choice == 'Rock':
    print("You win!")
  else:
    print("You lose!")
```
